```python
import requests

# Define the GraphQL endpoint
graphql_endpoint = 'https://countries.trevorblades.com/'

# Define your GraphQL query
graphql_query = '''
{
  countries {
    code
    name
    continent {
      name
    }
  }
}
'''

# Create a GraphQL request using the requests library
response = requests.post(graphql_endpoint, json={'query': graphql_query})

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    countries = data['data']['countries']

    # Print the results
    print("List of countries:")
    for country in countries:
        print(f"Country Code: {country['code']}")
        print(f"Country Name: {country['name']}")
        print(f"Continent: {country['continent']['name']}")
        print()
else:
    print(f"Error: {response.status_code}\n{response.text}")
```
